```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cars-1.csv')
```

```python
df.columns
```

1. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame.

```python
# Extract the 'cyl' and 'model' columns
cyl_values = df['cyl'].value_counts()
model_values = df['model'].value_counts()

# Create a figure and axes for the pie chart
fig, ax = plt.subplots()

# Plot the pie chart for 'cyl' column
ax.pie(cyl_values, labels=cyl_values.index, autopct='%1.1f%%')

# Set the aspect ratio to 'equal' for a circular pie chart
ax.axis('equal')

# Set the title
ax.set_title('Cylinders Distribution')

# Display the pie chart
plt.show()

# Create a figure and axes for the pie chart
fig, ax = plt.subplots()

# Plot the pie chart for 'model' column
ax.pie(model_values, labels=model_values.index, autopct='%1.1f%%')

# Set the aspect ratio to 'equal' for a circular pie chart
ax.axis('equal')

# Set the title
ax.set_title('Model Distribution')

# Display the pie chart
plt.show()
```
