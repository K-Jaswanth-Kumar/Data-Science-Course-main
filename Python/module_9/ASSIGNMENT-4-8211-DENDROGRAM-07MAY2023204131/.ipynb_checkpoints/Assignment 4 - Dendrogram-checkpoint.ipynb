```python
Data =[[5,3],
        [10,15],
        [15,12],
        [24,10],
        [30,30],
        [85,70],
        [71,80],
        [60,78],
        [70,55],
        [80,91]]
```

```python
import numpy as np
np_array = np.array(Data)
```

```python
import matplotlib.pyplot as plt

# Create scatter plot
plt.scatter(np_array[:,0], np_array[:,1])

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')

# Show the plot
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(np_array, method='ward')
dendrogram(Z)

# Set labels and title
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.title('Dendrogram')

# Show the plot
plt.show()
```
